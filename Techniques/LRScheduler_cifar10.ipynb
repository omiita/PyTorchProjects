```python
import glob
import os
import torch
import torchvision
import PIL
import time
```

```python
class CIFAR10Dataset(torch.utils.data.Dataset):
  def __init__(self, img_path, labels_idx_dict, transform=None):
    super(CIFAR10Dataset,self).__init__()
    self.img_path = img_path
    self.transform = transform
    self.labels_idx_dict = labels_idx_dict

  def __getitem__(self, idx):
    img_p = self.img_path[idx]
    img = PIL.Image.open(img_p)
    label = img_p.split('/')[-1].split('_')[-1].replace('.png','')
    label_idx = self.labels_idx_dict[label]

    if self.transform:
      img = self.transform(img)

    return img, label_idx

  def __len__(self):
    return len(self.img_path)
```

```python
num_epoch = 200
batch_size = 512
schedule = False
```

```python
if schedule:
  print('Scheduler!')
else:
  print('No Scheduler')

train_path = './cifar/train/*.png'
test_path = './cifar/test/*.png'

train_imgs = glob.glob(train_path)
test_imgs = glob.glob(test_path)
labels = []
with open('./cifar/labels.txt','r') as f:
  labels = f.readlines()
labels = [label.replace('\n','') for label in labels]
labels_idx_dict = {label:i for i, label in enumerate(labels)}


transform_dict = {
    'train':torchvision.transforms.Compose([
                                            torchvision.transforms.ToTensor(),
                                            torchvision.transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5])
    ]),
    'test':torchvision.transforms.Compose([
                                           torchvision.transforms.ToTensor(),
                                           torchvision.transforms.Normalize([.5,.5,.5],[.5,.5,.5])
    ])
}

train_ds = CIFAR10Dataset(train_imgs, labels_idx_dict, transform=transform_dict['train'])
test_ds = CIFAR10Dataset(test_imgs, labels_idx_dict, transform=transform_dict['test'])

train_dl = torch.utils.data.DataLoader(train_ds, batch_size=batch_size, shuffle=True)
test_dl = torch.utils.data.DataLoader(test_ds, batch_size=batch_size, shuffle=False)

net = torchvision.models.resnet50(pretrained=True)
net.fc = torch.nn.Linear(2048,10)

criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(),lr=1e-3,betas=(0.9,0.999))

device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
torch.backends.cudnn.benchmark = True
print("Device:",device)

if schedule:
  T_max = num_epoch
  scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer,T_max)

hist = {'loss':[], 'acc':[],'val_loss':[], 'val_acc':[]}

net.to(device)
print('START:')
net.train()
for i in range(num_epoch):
  since = time.time()

  epoch_loss = 0.
  epoch_corrects = 0.
  num_img = 0
  for batch in train_dl:
    imgs, labels = batch
    num_img += len(imgs)
    # print("{} / {}".format(num_img, len(train_ds)))
    imgs, labels = imgs.to(device), labels.to(device)
    

    optimizer.zero_grad()
    outputs = net(imgs)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()

    epoch_loss += loss.item() * imgs.size(0)
    # print(epoch_loss)
    epoch_corrects += sum(outputs.argmax(1) == labels.data)

  l = epoch_loss/num_img
  a = epoch_corrects/num_img
  hist['loss'].append(l)
  hist['acc'].append(a)

  dur = time.time() - since

  print("Epoch {}: Loss: {:.4f} Acc: {:.4f} Time: {}".format(i+1, l, a, dur))

  net.eval()
  val_imgs = 0.
  loss_ = 0.
  corrects = 0.
  for batch in test_dl:
    imgs, labels = batch
    val_imgs += len(imgs)
    imgs,labels = imgs.to(device), labels.to(device)

    with torch.no_grad():
      outputs = net(imgs)
      loss = criterion(outputs, labels)

      loss_ += loss.item()*imgs.size(0)
      corrects += sum(outputs.argmax(1) == labels.data)

  l = loss_ / val_imgs
  a = corrects / val_imgs
  hist['val_loss'].append(l)
  hist['val_acc'].append(a)

  print('Val: Loss {:.4f} Acc {:.4f}'.format(l, a))

  if schedule:
    scheduler.step()

if schedule:
  hist_sch = hist
else:
  hist_no_sch = hist


```

```python
import matplotlib.pyplot as plt
%matplotlib inline

epochs = range(1, num_epoch+1)

plt.title('Loss')
plt.plot(epochs, hist_sch['loss'],'ro',label='train CosAnneal')
plt.plot(epochs, hist_sch['val_loss'],'r',label='val CosAnneal')
plt.plot(epochs, hist_no_sch['loss'],'bo',label='train')
plt.plot(epochs, hist_no_sch['val_loss'],'b',label='val')

plt.legend()

plt.figure()
plt.title('Accuracy')
plt.plot(epochs, hist_sch['acc'],'ro',label='train CosAnneal')
plt.plot(epochs, hist_sch['val_acc'],'r',label='val CosAnneal')
plt.plot(epochs, hist_no_sch['acc'],'bo',label='train')
plt.plot(epochs, hist_no_sch['val_acc'],'b',label='val')
plt.ylim(0.7,1.)
plt.legend()
```
